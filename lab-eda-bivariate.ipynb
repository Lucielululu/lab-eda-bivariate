```python
# Part 1: Analyzing Best-Seller Trends Across Product Categories
# Objective: Understand the relationship between product categories and their best-seller status.
```

```python
# Crosstab Analysis: Create a crosstab between the product category and the isBestSeller status.
crosstab_result = pd.crosstab(df['category'], df['isBestSeller'])
crosstab_result
```

```python
# Crosstab Analysis:
# Create a crosstab between the product category and the isBestSeller status.
# Are there categories where being a best-seller is more prevalent?
# Hint: one option is to calculate the proportion of best-sellers for each category and then sort the categories based on this proportion in descending order.
# 创建一个新column用公式算出百分比，再用.sort_values排序。
crosstab_result['bestseller_proportion'] = (
    crosstab_result[True] / (crosstab_result[True] + crosstab_result[False])
)
crosstab_result.sort_values('bestseller_proportion', ascending=False)
```

```python
# Statistical Tests:

# Conduct a Chi-square test to determine if the best-seller distribution is independent of the product category.
from scipy.stats import chi2_contingency
chi2_statistic, chi2_p_value, _, _ = chi2_contingency(crosstab_result)
chi2_statistic, chi2_p_value
```

```python
# Compute Cramér's V to understand the strength of association between best-seller status and category.
from scipy.stats.contingency import association
association(crosstab_result, method="cramer")
```

```python
# Visualizations:
# Visualize the relationship between product categories and the best-seller status using a stacked bar chart.
# 太多category了看不出。
crosstab_result.head().plot(kind="bar", stacked=True)
plt.show()
```

```python
# Part 2: Exploring Product Prices and Ratings Across Categories and Brands
# Objective: Investigate how different product categories influence product prices.
```

```python
# Preliminary Step: Remove outliers in product prices.
# For this purpose, we can use the IQR (Interquartile Range) method. 
# Products priced below the first quartile minus 1.5 times the IQR or above the third quartile plus 1.5 times the IQR will be considered outliers and removed from the dataset. 
# The next steps will be done with the dataframe without outliers.
# Hint: you can check the last Check For Understanding at the end of the lesson EDA Bivariate Analysis for a hint on how to do this.

Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
    
# Define bounds for the outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
    
# Filter out outliers
df_clean = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]

df_clean
```

```python
# Violin Plots: 
# Use a violin plot to visualize the distribution of price across different product categories. 不同种类的价格的分布。
# Filter out the top 20 categories based on count for better visualization.
# Which product category tends to have the highest median price? Don't filter here by top categories.

# ✅ Step 1: Violin Plot for Top 20 Categories
# Count the top 20 most frequent categories 这些章节的各种建立新的 df 很烧脑。
top_20_categories = df_clean['category'].value_counts().nlargest(20).index

# Filter dataframe to include only those categories
df_top_20 = df_clean[df_clean['category'].isin(top_20_categories)]

# Set up the plot
plt.figure(figsize=(15, 8))
sns.violinplot(data=df_top_20, x='category', y='price', inner='box')
# 🔍 inner='box' adds a boxplot inside each violin for better median/quantile visibility.

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right')
plt.title('Price Distribution Across Top 20 Product Categories')
plt.tight_layout()
plt.show()

# ✅ Step 2: Find Category with Highest Median Price (All Categories)

# Group by category and calculate median price
median_prices = df_clean.groupby('category')['price'].median()

# Sort and find the one with the highest median
highest_median_category = median_prices.idxmax()
highest_median_value = median_prices.max()

print(f"The category with the highest median price is: '{highest_median_category}' with a median of ${highest_median_value:.2f}")
```

```python
# Bar Charts:
# Create a bar chart comparing the average price of products for the top 10 product categories (based on count).
# Which product category commands the highest average price? Don't filter here by top categories.
```

```python
# ✅ Step 1: Bar charts for Top 10 Categories
# create new df with top 10 categories
top_10_categories = df_clean['category'].value_counts().nlargest(10).index #不明白这行code。

# Filter dataframe to include only those categories
df_top_10 = df_clean[df_clean['category'].isin(top_10_categories)]

# Set up the plot
plt.figure(figsize = (15,8))
sns.barplot(data = df_top_10, x='category', y='price')
plt.xticks(rotation=45, ha='right')
plt.title('Price Distribution Across Top 10 Product Categories')
plt.tight_layout()
plt.show()

# ✅ Step 2: Find Category with Highest Average Price (All Categories)

# Group by category and calculate mean (average) price
mean_prices = df_clean.groupby('category')['price'].mean()

# Sort and find the one with the highest median
highest_mean_category = mean_prices.idxmax()
highest_mean_value = mean_prices.max()

print(f"The category with the highest mean price is: '{highest_mean_category}' with a mean of ${highest_mean_value:.2f}")
```

```python
# Box Plots:
# Visualize the distribution of product ratings based on their category using side-by-side box plots. 
# Filter out the top 10 categories based on count for better visualization.
# Which category tends to receive the highest median rating from customers? Don't filter here by top categories.
```

```python
# ✅ Step 1: Bar charts for Top 10 Categories
# create new df with top 10 categories (by rating)
top_10_categories_rating = df_clean['stars'].value_counts().nlargest(10).index #不明白这行code。

# Filter dataframe to include only those categories
df_top_10_rating = df_clean[df_clean['stars'].isin(top_10_categories_rating)]

# Set up the plot
plt.figure(figsize = (15,8))
sns.boxplot(data = df_top_10_rating, x='stars', y='price')
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Product Categories by Rating')
plt.tight_layout()
plt.show()


# ✅ Step 2: Find Category with Highest median rating (All Categories)

# Group by category and calculate mean (average) price
median_rating = df_clean.groupby('category')['stars'].median()

# Sort and find the one with the highest median
highest_median_rating = median_rating.idxmax()
highest_median_rating_value = median_rating.max()

print(f"The category with the highest mean price is: '{highest_median_rating}' with a median of {highest_median_rating_value:.1f} stars.")
```

Part 3: Investigating the Interplay Between Product Prices and Ratings
Objective: Analyze how product ratings (stars) correlate with product prices.

Correlation Coefficients:

Calculate the correlation coefficient between price and stars.
Is there a significant correlation between product price and its rating?
Visualizations:

Use a scatter plot to visualize the relationship between product rating and price. What patterns can you observe?
Use a correlation heatmap to visualize correlations between all numerical variables.
Examine if product prices typically follow a normal distribution using a QQ plot.

```python
# Part 3: Investigating the Interplay Between Product Prices and Ratings Objective: 
# Analyze how product ratings (stars) correlate with product prices.


# Step 1 : Conduct Correlation Coefficients Analysis to calculate whether there is a significant correlation between product price and its rating.
from scipy.stats import pearsonr

# Apply Pearson Correlation
correlation, p_value = pearsonr(df_clean['price'], df_clean['stars'])

print(f"Correlation coefficient: {correlation:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("The correlation is statistically significant")
else:
    print("The correlation is not statistically significant")
    
    
# Step 2: Scatter plot: to visualize the relationship between product rating and price. 
# Create figure
plt.figure(figsize=(10, 6))

# Scatter plot with trend line
sns.regplot(x='stars', y='price', data=df, 
            scatter_kws={'alpha':0.4, 'color':'blue'},
            line_kws={'color':'red', 'linewidth':2})

# Add labels and title
plt.title('Amazon Product Ratings vs. Price', fontsize=14)
plt.xlabel('Price ($)', fontsize=12)
plt.ylabel('Rating (Stars)', fontsize=12)
plt.ylim(0, 5.5)  # Ratings typically 1-5 stars

# Add correlation text
corr = df_clean['price'].corr(df_clean['stars'])
plt.text(0.7, 0.95, f'Correlation: {corr:.2f}', 
         transform=plt.gca().transAxes, fontsize=12)

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Use a correlation heatmap to visualize correlations between all numerical variables. 
# Calculate correlations
corr = df_clean.select_dtypes(include=['number']).corr()

# Create heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr, 
            annot=True, 
            fmt=".2f", 
            cmap='coolwarm', 
            center=0,
            linewidths=0.5,
            annot_kws={"size": 10})

# Customize plot
plt.title('Correlation Heatmap of Amazon Product Metrics', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Examine if product prices typically follow a normal distribution using a QQ plot.

from scipy import stats
# Create figure
plt.figure(figsize=(10, 6))

# Generate QQ plot
stats.probplot(df_clean['price'], dist="norm", plot=plt)

# Add title and labels
plt.title('QQ Plot - Testing Normality of Product Prices', fontsize=14)
plt.xlabel('Theoretical Quantiles', fontsize=12)
plt.ylabel('Ordered Price Values', fontsize=12)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
